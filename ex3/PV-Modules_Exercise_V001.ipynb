```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#help(pd.read_csv)
```

```python
# plotstyle = "*-" # default color blue), asterics marker, continuous line
plotstyle = "ro-" # red, filled circle marker, continuous line
```

```python
# load data
df_single = pd.read_csv("./data/I-V_single_module.txt", comment = "#", names = ["V", "ImA", "R"])
df_single.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,4), dpi = 136)

ax[0].plot(df_single["V"], df_single["ImA"], plotstyle) 

ax[0].set_title("I-V Curve of a Single PV Mini Module")
ax[0].set_xlabel("Voltage $V$ in V")
ax[0].set_ylabel("Current $I$ in mA")
ax[0].grid(True)

ax[0].set_xlim(0,6)
ax[0].set_ylim(0,22)

ax[0].set_xticks(np.arange(0,6.01,0.5)) 
ax[0].set_yticks(np.arange(0,22.01,2)) 

ax[1].plot(df_single["V"], df_single["ImA"]*df_single["V"], plotstyle) 

ax[1].set_title("Power Curve of a Single PV Mini Module")
ax[1].set_xlabel("Voltage $V$ in V")
ax[1].set_ylabel("Power $P$ in mW")
ax[1].grid(True)

ax[1].set_xlim(0,6)
ax[1].set_ylim(0,100)

ax[1].set_xticks(np.arange(0,6.01,0.5)) 
ax[1].set_yticks(np.arange(0,100.01,10)) 

plt.plot()
```
